```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    books = []

    # Each book title is inside <h3>
    items = soup.find_all('h3')

    for item in items:
        title = item.text.strip()

        # Move to next elements
        parent = item.find_parent()

        text_block = parent.get_text(separator=" ").strip()

        import re

        # Publisher (before "| Magazines")
        pub_match = re.search(r'\n(.*?)\s*\|\s*Religious', text_block)
        publisher = pub_match.group(1).strip() if pub_match else None

        # Published
        published_match = re.search(r'Published:\s*(.*)', text_block)
        published = published_match.group(1).strip() if published_match else None

        # Downloads
        downloads_match = re.search(r'Downloads:\s*(\d+)', text_block)
        downloads = downloads_match.group(1) if downloads_match else None

        # Pages
        pages_match = re.search(r'Pages:\s*(.*)', text_block)
        pages = pages_match.group(1).strip() if pages_match else None

        books.append({
            "Title": title,
            "Publisher": publisher,
            "Published": published,
            "Downloads": downloads,
            "Pages": pages,
            "type" : "Religious"
        })

    return books
```

```python
poetry_data = []

for page in range(1, 84):
    if page == 1:
        url = "https://www.free-ebooks.net/religious"
    else:
        url = f"https://www.free-ebooks.net/religious/{page}"

    print(f"Scraping page {page}...")

    data = scrape_page(url)

    print(f"Found {len(data)} records")

    poetry_data.extend(data)

    time.sleep(2)
```

```python
# Use 'latin1' or 'cp1252' to handle the 0xa0 character
df_existing = pd.read_csv("Metadata_Ebooks.csv", encoding='latin1') 

df_poetry = pd.DataFrame(poetry_data)

# Combine, clean, and save
df_final = pd.concat([df_existing, df_poetry], ignore_index=True)
df_final.drop_duplicates(inplace=True)
df_final.to_csv("Metadata_Ebooks.csv", index=False)
```
